## How do we select the date?
1. If Date is given - Date
2. If Date is crap - New_Date_from_Inv
3. else Year
4. else drop the row (np.nan)

```python
import datetime
def clean_date(x):
    
    """
        This function takes a row of DF as argument
    """
    
    if len(x.Date.split('-'))==3:
        x=[i for i in x.Date.split() if i.count('-')==2][0]
        x1='%y' if x[-3]=='-' else '%Y'
        try:
            return datetime.datetime.strptime(x, f"%d-%b-{x1}").date()
        except ValueError:
            print(x)
            
    elif len(x.New_Date_from_Inv.split('/'))==3:
        return x.New_Date_from_Inv
    elif x.Year>0:
        return x.Year
    else:
        return np.nan
```

```python
df[['Date','Year','New_Date_from_Inv']].apply(clean_date, axis=1)
```

```python
loa=['swim','surf','fish','div','bath','boat','snork','board','sail','wad','kayak','float']
```

```python
def func(x):
    if type(x)==float: # if it was missing it remain missing
        return np.nan
    for i in loa:     
        if i in x:    # if the value from the list is in activity
            return i
    return 'other'    # if non of activities are in the value then other
```

```python
df.Activity=df.Activity.str.lower().map(func)
```

```python
def mod(x):
    """x is a Serie
    
    """
    
    if x.isna().sum()==x.shape[0]:
        return np.nan
    else:
        return x.mode()[0]
```

```python
df_apc=df[['Country','Activity']].groupby('Country').agg(mod)
```
