# Cleaning

* Missing Values
    * Detect them by counting them
    * Null them
    * Interpolate
    * Drop them
* Inconsistant data (Logical errors)
* Outliers
    * Detect using IQR (inter-quartile range), Z-score (normal distribution based approach)
    * Drop them
    * Fix them
    * Trancate them (Winsorizing)
    * Leave them as they are
* Text error
    * Special Characters é è ê ó ü ä 
    * Numbers that are considered as text
    * Empty characters
    * Different notations Ave. des Champs d'Elysees and Avenue des Champs d'Elysees, PSG 
    * Mixed case (upper and lower), punctuation marks (!,'".:;etc)
    * Special Solutions - NLP (Natural Language Processing)
* Duplicates
* Low variance

# Outliers

```python
df.describe().T;
```

```python
df.fuelCost08.plot.box();
```

```python
stats=df.describe().transpose()
```

```python
stats['IQR']=stats['75%']-stats['25%']
```

```python
outliers=pd.DataFrame(columns=df.columns)
```

```python
outliers
```

```python
for col in stats.index:
    iqr=stats.IQR[col]
    cutoff=iqr*1.6
    lower=stats.at[col,'25%']-cutoff
    upper=stats.at[col,'75%']+cutoff
    results=df[(df[col]<lower)|(df[col]>upper)].copy()
    results['Outlier']=col
    outliers=outliers.append(results)
```
