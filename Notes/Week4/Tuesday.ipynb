```python
import numpy as np
```

```python
x1=np.arange(0,10,1)
```

```python
x2=np.linspace(0,5,10)
```

```python
plt.plot(x1,x1,label='linear')
plt.plot(x1,x2**2,label='quadratic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title('Linear vs Quadratic plots')
plt.legend()
plt.show()
```

```python
fig,[ax1,ax2]=plt.subplots(1,2)

ax1.plot(x1,x1,label='linear')
ax2.plot(x1,x2**2,label='quadratic')

ax1.set_xlabel('x label')
ax1.set_ylabel('y label')
ax1.set_title('Linear')

ax2.set_xlabel('x label')
ax2.set_title('Quadratic')

plt.show()
```

```python
plt.scatter(x1,x2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
labels=['Vegs','Vegans','Non-Vegs']
sizes=[40,45,15]
```

```python
explode=(0,0.1,0)
fig,ax=plt.subplots()
ax.pie(sizes,explode=explode,labels=labels,shadow=True, startangle=90,autopct='%1.1f%%')
ax.axis('equal')
plt.show()
```
