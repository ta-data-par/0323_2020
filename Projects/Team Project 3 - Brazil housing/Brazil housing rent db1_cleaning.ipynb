```python
#We can see that there we can just need to get rid of the sign for those columns
```

```python
lst=['hoa','rent_amount','property_tax','fire_insurance','total']
for i in lst:
    df[i]=df[i].apply(lambda x: x.strip('R$').replace('Sem info','0').replace('Incluso','0').replace(',','').replace(' ',''))
    df[i]=df[i].astype(float)
```

```python
print(df.floor.mode())
df.floor=df.floor.apply(lambda x: x.replace('-','0'))
df.floor=df.floor.astype(float)
```
